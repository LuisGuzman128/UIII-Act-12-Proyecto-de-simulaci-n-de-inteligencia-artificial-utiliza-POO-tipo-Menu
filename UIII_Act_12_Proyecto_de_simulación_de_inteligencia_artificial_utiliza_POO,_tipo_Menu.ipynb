```python
# 23/mayo/2026

# INGENIERO:
# Rama de ingeniería, proyectos firmados,
# colegiación, software de cálculo,
# presupuesto de obra y experiencia en campo.

# ======================================================
# 🏗️ CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Manuel Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 27
# 🏗️ Proyecto: Simulación de Inteligencia Artificial
# ======================================================

# =========================
# 🏗️ Clase Ingeniero
# =========================
class Ingeniero:

    # -----------------------------------
    # Constructor
    # -----------------------------------

    def __init__(self):

        self.ingenieros = {}

    # -----------------------------------
    # Agregar ingeniero
    # -----------------------------------

    def agregar_ingeniero(self):

        clave = input("Ingresa clave del ingeniero: ")
        rama = input("Ingresa rama de ingeniería: ")
        proyectos = input("Ingresa proyectos firmados: ")
        colegiacion = input("Ingresa colegiación: ")
        software = input("Ingresa software de cálculo: ")
        presupuesto = input("Ingresa presupuesto de obra: ")
        experiencia = input("Ingresa experiencia en campo: ")

        self.ingenieros[clave] = {

            "Rama": rama,
            "Proyectos": proyectos,
            "Colegiacion": colegiacion,
            "Software": software,
            "Presupuesto": presupuesto,
            "Experiencia": experiencia
        }

        print("\n✅ Ingeniero agregado correctamente")

    # -----------------------------------
    # Listar ingenieros
    # -----------------------------------

    def listar_ingenieros(self):

        print("\n===================================================================================================================================================")
        print(f"{'CLAVE':<12}{'RAMA':<20}{'PROYECTOS':<22}{'COLEGIACION':<18}{'SOFTWARE':<20}{'PRESUPUESTO':<20}{'EXPERIENCIA':<20}")
        print("===================================================================================================================================================")

        for key, value in self.ingenieros.items():

            print(f"{key:<12}"
                  f"{value['Rama']:<20}"
                  f"{value['Proyectos']:<22}"
                  f"{value['Colegiacion']:<18}"
                  f"{value['Software']:<20}"
                  f"{value['Presupuesto']:<20}"
                  f"{value['Experiencia']:<20}")

        print("===================================================================================================================================================")

    # -----------------------------------
    # Actualizar ingeniero
    # -----------------------------------

    def actualizar_ingeniero(self):

        actualizar = input("Ingresa clave del ingeniero: ")

        if actualizar in self.ingenieros:

            rama = input("Nueva rama de ingeniería: ")
            proyectos = input("Nuevos proyectos firmados: ")
            colegiacion = input("Nueva colegiación: ")
            software = input("Nuevo software de cálculo: ")
            presupuesto = input("Nuevo presupuesto de obra: ")
            experiencia = input("Nueva experiencia en campo: ")

            self.ingenieros[actualizar] = {

                "Rama": rama,
                "Proyectos": proyectos,
                "Colegiacion": colegiacion,
                "Software": software,
                "Presupuesto": presupuesto,
                "Experiencia": experiencia
            }

            print("\n✅ Ingeniero actualizado correctamente")

        else:
            print("❌ Ingeniero no encontrado")

    # -----------------------------------
    # Borrar ingeniero
    # -----------------------------------

    def borrar_ingeniero(self):

        borrar = input("Ingresa clave del ingeniero a borrar: ")

        if borrar in self.ingenieros:

            del self.ingenieros[borrar]

            print("\n✅ Ingeniero eliminado correctamente")

        else:
            print("❌ Ingeniero no encontrado")


# =========================
# 🏗️ Objeto principal
# =========================

sistema = Ingeniero()

# =========================
# 📌 Menú principal
# =========================

while True:

    print("\n=========== MENÚ INGENIERO 🏗️ ===========")
    print("1. Agregar ingeniero")
    print("2. Listar ingenieros")
    print("3. Actualizar ingeniero")
    print("4. Borrar ingeniero")
    print("5. Salir")

    opcion = int(input("Selecciona una opción: "))

    # -----------------------------------
    # OPCIÓN 1
    # -----------------------------------

    if opcion == 1:

        sistema.agregar_ingeniero()

    # -----------------------------------
    # OPCIÓN 2
    # -----------------------------------

    elif opcion == 2:

        sistema.listar_ingenieros()

    # -----------------------------------
    # OPCIÓN 3
    # -----------------------------------

    elif opcion == 3:

        sistema.actualizar_ingeniero()

    # -----------------------------------
    # OPCIÓN 4
    # -----------------------------------

    elif opcion == 4:

        sistema.borrar_ingeniero()

    # -----------------------------------
    # OPCIÓN 5
    # -----------------------------------

    elif opcion == 5:

        print("\n🏗️ Programa finalizado")
        break

    # -----------------------------------
    # OPCIÓN INVÁLIDA
    # -----------------------------------

    else:

        print("❌ Opción no válida")
```
